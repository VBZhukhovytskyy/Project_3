```python
# Read CSV_4
alcohol_consumption_2017 = "Resources/Alcohol consumption_2017.csv"
alcohol_consumption_2017_df = pd.read_csv(alcohol_consumption_2017)
alcohol_consumption_2017_df.head(10)
```

```python
# Merge two data frames together
alcohol_suicide_df = pd.merge(suicide_2017_df,alcohol_consumption_2017_df, how='outer', on='Country')
alcohol_suicide_df.head(10)                       
```

```python
lit_alc = alcohol_consumption_2017_df.iloc[:, 2]
lit_alc.head(10)
```

```python
m_suicide = suicide_2017_df.iloc[:, 2]
m_suicide.head(10)
```

```python
vc_slope, vc_int, vc_r, vc_p, vc_std_err = stats.linregress(
    lit_alc, m_suicide)
vc_fit = vc_slope * lit_alc + vc_int
```

```python
# Plot
fig,(ax1) = plt.subplots(1, sharex=True)
fig.suptitle("Alcohol-Suicide(M)", fontsize=16, fontweight="bold")

ax1.set_xlim(min(lit_alc), max(lit_alc))
ax1.plot(lit_alc, m_suicide, linewidth=0, marker="o")
ax1.plot(lit_alc, vc_fit, linewidth=1)
ax1.set_ylabel("Suiside per 100k (male)",fontsize = 16, fontweight="bold")
ax1.set_xlabel("Alcohol Consumption, L of pure alcohol",fontsize = 16, fontweight="bold")

plt.savefig("../Project_3/final-plot.png")
```

```python
# Print results and save image
print("The suicide rate for male population 2017 equals" +
      str(vc_slope * lit_alc + vc_int) + ".")
```

```python
xs = np.array([6.035,4.532,7.745,6.256,8.929,5.029,6.895,7.023,7.032,6.882,3.009,4.051,6.415,8.801,6.289,5.999,9.045,6.302,4.418,5.808],
dtype=np.float64)
ys = np.array([13.6 , 13.9, 21.7, 15.8, 20.8, 12.9, 15.1, 17.3, 17.4, 15.8, 8.20, 12.8, 17.5, 21.1, 17.6, 13.6, 22.2, 15, 11.9, 16], dtype=np.float64)

def best_fit_slope_and_intercept(xs,ys):
    m = (((mean(xs)*mean(ys)) - mean(xs*ys)) /
         ((mean(xs)*mean(xs)) - mean(xs*xs)))
    b = mean(ys) - m*mean(xs)
    return m, b

def squared_error(ys_orig,ys_line):
    return sum((ys_line - ys_orig) * (ys_line - ys_orig))

def coefficient_of_determination(ys_orig,ys_line):
    y_mean_line = [mean(ys_orig) for y in ys_orig]
    squared_error_regr = squared_error(ys_orig, ys_line)
    squared_error_y_mean = squared_error(ys_orig, y_mean_line)
    return 1 - (squared_error_regr/squared_error_y_mean)
    
m, b = best_fit_slope_and_intercept(xs,ys)
regression_line = [(m*x)+b for x in xs]

r_squared = coefficient_of_determination(ys,regression_line)
print("Coefficient of determination (r-squared) equals",r_squared)
```
